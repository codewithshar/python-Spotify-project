```python
#set index as date
Spotify_data=pd.read_csv('D:/spotify/tracks.csv',index_col='release_date')
Spotify_data
```

```python
#if we want specific value from data set 
Spotify_data[["artists"]].iloc[18]
```

```python
# convert songs timming"milliseconds" to "seconds"
Spotify_data["duration"]=Spotify_data["duration_ms"].apply(lambda x: round(x/1000))
Spotify_data.drop("duration_ms", inplace=True, axis=1)
```

```python
Spotify_data.duration.head()
#all will be converted into seconds
```

```python
#co-relation map
```

```python
cor_df= Spotify_data.drop(["key","mode","explicit"],axis=1).corr(method ="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(cor_df ,annot=True,fmt=".1g",vmin=1,vmax=1,center=0, cmap="inferno",linewidth=1, linecolor='Black')
heatmap.set_title("Correlation Heatmap between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation = 90)
```

```python
sample_df = Spotify_data.sample(int(0.004*len(Spotify_data)))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data= sample_df, y= "loudness", x ="energy", color = "c").set(title="loudness and energy correltion")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data= sample_df, y= "popularity", x ="acousticness", color = "b").set(title="popularity and acousticness correltion")
```
